### Fit the linear model and predict per capita income of canada for the year 2020

```python
#IMPORTING LIABRARIES
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# LOADING DATASETS
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df
```

```python
# STATISTICS OF THE DATA
```

```python
df.describe()
```

```python
# CHECKING FOR NULL VALUES 
```

```python
df.isnull()
```

```python
# VISUALIZATION
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('pci')
plt.scatter(df.year,df.per_capita_income,color='red',marker='*')
```

```python
#2D ARRAY
```

```python
new_df=df.drop('per_capita_income',axis='columns')
new_df
```

```python
# MODEL FITTING
```

```python
model=linear_model.LinearRegression()
model.fit(new_df,df.per_capita_income)
```

```python
# MAKING PREDICTION
```

```python
model.predict([[2020]])
```
